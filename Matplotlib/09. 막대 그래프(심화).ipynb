# 9. 막대 그래프(심화)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False

label = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]

plt.bar(label, values)
```

```python
plt.barh(label, values) #barh: 가로 막대 그래프
```

```python
plt.barh(label, values)
plt.xlim(175, 195) # x축 범위 제한ㅠㅁㄱ
```

```python
bar = plt.bar(label, values)
bar[0].set_hatch('/') # ////
bar[1].set_hatch('x') # xxxx
bar[2].set_hatch('..') # ... 구글에 matplotlib hatch 검색 후 종류 확인 가능
```

```python
bar = plt.bar(label, values)
plt.ylim(175, 195)

# idx는 index로 0, 1, 2 순서로 대임, rect는 각 bar의 높이 대로 순서대로 대입, enumerate(bar)는 첫줄의 bar로부터 정보 가져옴
for idx, rect in enumerate(bar):
    plt.text(idx, rect.get_height() + 0.5, values[idx], ha='center', color='b') # plt.text(x좌표, y좌표, 값)
#rect: bar에서 갖는 좌표, 너비 및 높이를 갖는 인자
```
