# 2. 축

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
#plt.title('꺾은선 그래프', fontdict={'family':'HYGungSo-Bold', 'size':20}) # 개별 폰트 설정 - 타이틀만 궁서채로 바뀜
```

```python
plt.plot(x,y)
plt.title('꺾은선 그래프')
```

```python
plt.plot(x,y)
plt.xlabel('x축')
plt.ylabel('y축')
```

```python
plt.plot(x,y)
plt.xlabel('x축', color='red')
plt.ylabel('y축', color='#00aa00') # RGB로 표현
```

```python
plt.plot(x,y)
plt.xlabel('x축', color='red', loc='right') # left, right, center
plt.ylabel('y축', color='#00aa00', loc='top') # top, center, bottom
```

```python
plt.plot(x,y)
plt.xticks([1, 2, 3])
plt.yticks([3, 6, 9, 12]) # 원하는 틱만 나타낼 수 있음
plt.show()
```
