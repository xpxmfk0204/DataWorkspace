# 3. 범례(legend)
그래프에서 데이터가 어떤걸 의미하는지 알려주는 역할

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False

x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x,y, label = '무슨 데이터')
plt.legend()
```

```python
plt.plot(x,y, label = '무슨 데이터')
plt.legend(loc='upper right')
```

```python
plt.plot(x,y, label = '무슨 데이터')
plt.legend(loc='lower right') # 구글에 matplotlib legend 검색해 공식 홈페이지 들어가면 location string 확인 가능 best는 가장 적절한 곳
```

```python
plt.plot(x,y, label = '무슨 데이터')
plt.legend(loc=(0.5, 0.5)) # x축, y축 기준으로(0~1사이)
```
