# 13. 원 그래프

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']

plt.pie(values, labels=labels, autopct='%.1f') # plt.pie(값, 이름, ...), autopct='%.1f': 퍼센트로 나타내기
plt.show() # 그래프의 시작은 '3시'방향부터 반시계 방향으로 진행
```

```python
values = [1, 1, 1, 1, 1, 1] # 각 value를 100점으로 환산해서 나오는 비율을 나타내는 것
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']

plt.pie(values, labels=labels, autopct='%.1f')
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']

plt.pie(values, labels=labels, autopct='%.1f%%') # 비율에 %표시하기
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']

plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False) #그래프 시작 위치 변경, 시계방향으로 진행
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']
explode = [0.2, 0, 0.1, 0, 0, 0] # 강조하고 싶은 부분을 띄워서 표시

plt.pie(values, labels=labels, explode=explode)
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']
explode = [0.05]*6 # 모든 요소를 동일한 간격으로 띄우기

plt.pie(values, labels=labels, explode=explode)
plt.show()
```

```python
plt.pie(values, labels=labels, explode=explode)
#plt.legend(ncol=6)
plt.legend(loc=(1.2, 0.3)) # 범례 설정 및 위치 조정 
```

```python
plt.pie(values, labels=labels, explode=explode)
#plt.legend(ncol=6)
plt.legend(loc=(1.2, 0.3), title='언어별 선호도') # 범례에 제목 만들기
```

```python
plt.pie(values, labels=labels, explode=explode)
#plt.legend(ncol=6)
plt.legend(loc=(1.2, 0.3))
plt.title('언어별 선호도') # 그래프에 제목 만들기
```
