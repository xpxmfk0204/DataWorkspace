# 4. 스타일

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False

x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, linewidth=5) # 선 굵기 조정
```

## 마커 Marker

```python
plt.plot(x,y, marker='o') # 데이터가 있는 부분 표시
```

```python
plt.plot(x,y, marker='o', linestyle='None') # 선 제거
```

```python
plt.plot(x,y, marker='v', markersize=10) # marker의 가양한 형태:구글에 matplotlib marker 검색해 홈페이지 들어가서 확인
```

```python
plt.plot(x,y, marker='X', markersize=10)
```

```python
plt.plot(x,y, marker='v', markersize=10, markeredgecolor='red') # marker 가장자리 색 변경
```

```python
plt.plot(x,y, marker='v', markersize=10, markeredgecolor='red', markerfacecolor='yellow') # marker 색 변경
```

## 선 스타일

```python
plt.plot(x,y, linestyle = ':') # 구글에서 matplotlib linestyle 검색 후 홈페이지에서 확인
```

```python
plt.plot(x,y, linestyle = '--')
```

```python
plt.plot(x,y, linestyle = '-.')
```

## 색깔

```python
plt.plot(x,y, color='pink')
```

```python
plt.plot(x,y, color='#ff0000') # RGB로 표현
```

```python
plt.plot(x,y, color='b') # matplotlib에서 정의해둔 색, 구글에 matplotlib color 검색
```

```python
plt.plot(x,y, color='g')
```

## 포멧

```python
plt.plot(x,y, 'ro--') # color, marker, linestyle순으로 설정 'red', 'o', '--'
```

```python
plt.plot(x, y, 'bv:')
```

```python
plt.plot(x,y, 'go') # linestyle을 비우면 None과 같은 의미, 실선 아님
```

## 축약어

```python
plt.plot(x,y, marker='o', mfc='red', ms=10, mec='blue', ls=':')
# markerfacecolor, markersize, markeredgecolor, linestyle 축약
#구글에서 matplotlib markerfacecolor mfc 검색 후 축약어 확인
```

## 투명도

```python
plt.plot(x,y, marker='o', mfc='red', ms=10, alpha = 0.3) # alpha: 투명도(0~1), 강조할 것은 1로, 그렇지 않은 것은 얕게 정리
```

## 그래프 크기

```python
plt.figure(figsize=(10,5)) # 그래프의 width, height 설정
plt.plot(x,y)
```

```python
plt.figure(figsize=(5,10))
plt.plot(x,y)
```

```python
plt.figure(figsize=(10,5), dpi=200) #dots per inch(해상도)
plt.plot(x,y) # size는 10,5로 동일하지만 해상도가 높아 확대한 것처럼 커짐
```

## 배경색

```python
plt.figure(facecolor = 'yellow')
plt.plot(x,y)
```

```python
plt.figure(facecolor = '#a1c3ff') #color picker 검색해서 색 정보 가져올 수 있음
plt.plot(x,y)
```
