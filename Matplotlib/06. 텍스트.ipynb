# 6. 텍스트

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False

x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x,y, marker='o')

# enumerate(=세다, count) -> enumerate(y): y자료를 for문에 따라 차례로 세는 것, range 또는 'for i in [1, 2, 3]'와 비슷하게 생각
for idx, txt in enumerate(y):# y데이터 (2, 4, 8)를 그래프 위에 찍어주는 함수 txt는 2, 4, 8로 입력, idx는 0, 1, 2로 들어감
    plt.text(x[idx], y[idx], txt)
```

```python
plt.plot(x,y, marker='o')
for idx, txt in enumerate(y):
    plt.text(x[idx], y[idx]+0.3, txt, ha='center', color='blue') # ha(horizontal alignment): 수평 정렬
```
