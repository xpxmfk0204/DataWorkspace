# 15. 산점도 그래프
두 그래프의 상관관계 표시

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False

import pandas as pd
df = pd.read_excel('../Pandas/score.xlsx')
df
```

```python
df['학년'] =[3, 3, 2, 1, 1, 3, 2, 2] # 학년 column 생성, 학년 정보에 따라 산점도 스타일을 변경하기 위함
df
```

```python
plt.scatter(df['영어'], df['수학']) # 산점도 그래프 형성
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
import numpy as np
sizes = np.random.rand(8) * 1000
sizes
```

```python
plt.scatter(df['영어'], df['수학'], s=sizes) # 각 수치별 램덤의 마킹 크기 설정
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
sizes = df['학년']*500 # 1학년 = 500, 2학년 = 1000, 3학년 = 1500
plt.scatter(df['영어'], df['수학'], s=sizes)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
plt.scatter(df['영어'], df['수학'], s=sizes, c=df['학년'], cmap='viridis', alpha=0.4) 
# c: 색깔 분포 정의, cmap: color map 이런게 있다 정도만 확인
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
# 구글에 matplotlib cmap 검색 후 공홈에서 색 지정 가능(viridis도 그중 일부이며 스펙트럼에 따라 값에 해당하는 지점의 색 배정)
```

```python
plt.figure(figsize=(10, 10)) # 그래프 크기 수정
plt.scatter(df['영어'], df['수학'], s=sizes, c=df['학년'], cmap='viridis', alpha=0.4) 
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
plt.colorbar() # 색깔 별 데이터 정보 확인 가능
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(df['영어'], df['수학'], s=sizes, c=df['학년'], cmap='viridis', alpha=0.4) 
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
plt.colorbar(ticks=[1, 2, 3], label='학년', shrink=0.5) # label: colorbar에 제목 형성, shrink: colorbar 크기 조정
# 위의 1.25, 2.5 등의 불필요한 요소 제거 및 필요한 수치만 형성
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(df['영어'], df['수학'], s=sizes, c=df['학년'], cmap='viridis', alpha=0.4) 
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
plt.colorbar(ticks=[1, 2, 3], label='학년', shrink=0.5, orientation='horizontal')
# colorbar의 위치 조정 - horizontal: 수평 위치로
```
