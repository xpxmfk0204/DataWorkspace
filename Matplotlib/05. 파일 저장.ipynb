# 5. 파일 저장

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False

x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x,y)
plt.savefig('graph.png', dpi=100) # 저장 후 사이즈 키우기, 여기선 작은 크기 유지
```

```python
plt.figure(dpi=200)
plt.plot(x,y)
plt.savefig('graph_200.png') # 사이즈 키운 후 저장
```

```python
plt.figure(dpi=200)
plt.plot(x,y)
plt.savefig('graph_200.png', dpi=100) # 사이즈 키운 후 원하는 크기로 저장
```
