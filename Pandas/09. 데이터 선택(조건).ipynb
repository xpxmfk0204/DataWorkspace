# 9. 데이터 선택(조건)
조건에 해당하는 데이터 선택

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df['키'] >= 185 # 학생들의 키가 185 이상인지 여부를 True/False로
```

```python
fillt = (df['키'] >= 185)
df[fillt] # 키가 185 이상인 학생들의 정보만 불러오기
```

```python
df[~fillt] #fillter를 역으로 적용
```

```python
df[df['키']>= 185]
```

```python
df.loc[df['키'] >= 185, '수학'] # 키가 185 이상인 학생들의 수학 데이터
```

```python
df.loc[df['키'] >= 185, ['이름', '수학', '과학']] #키가 185 이상인 학생들의 이름, 수학, 과학 데이터
```

# 다양한 조건

### & 그리고

```python
df.loc[(df['키'] >= 185) & (df['학교']=='북산고')] # 키가 185 이상이고 북산고를 다니는 학생 데이터
```

### 또는 |

```python
df.loc[(df['키'] < 170) | (df['키'] > 200)] #키가 170보다 작거나 200보다 큰 학생 데이터
```

### str 함수

```python
fillt = df['이름'].str.startswith('송')
fillt
```

```python
fillt = df['이름'].str.startswith('송') # 이름이 '송'으로 시작하는 학생 데이터
df[fillt]
```

```python
fillt = df['이름'].str.contains('태') # 이름에 '태'가 들어가는 학생 데이터
df[fillt]
```

```python
df[~fillt] # 이름에 '태'가 들어가는 사람을 제외
```

```python
langs = ['Python', 'Java']
fillt = df["SW특기"].isin(langs) # 'SW특기'가 python, java인 학생 데이터
df[fillt]
```

```python
df
```

```python
langs = ['python', 'java']
fillt = df['SW특기'].str.lower().isin(langs) #소문자로 만들어서 찾기
df[fillt]
```

```python
fillt = df['SW특기'].str.contains('Java')
df[fillt]
```

```python
df['SW특기'].str.lower() # 4, 5번 데이터는 NaN데이터가 뜸
```

```python
df['SW특기'].str.lower().isin(langs) # True, False로 나타나야 필요한 것은 보여주고 필요없는 것은 안보여줄 수 있음
```

```python
df['SW특기'].str.contains('Java') # 4, 5번이 NaN데이터로 나옴 그래서 처리 못함
```

```python
df['SW특기'].str.contains('Java', na=True) # NaN 데이터에 대해서 True로 설정
```

```python
fillt = df['SW특기'].str.contains('Java', na=False) # NaN 데이터에 대해서 False로 설정
df[fillt]
```
